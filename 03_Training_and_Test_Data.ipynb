# Training and Test Data

## Importing the Libraries

```
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
from sklearn.model_selection import train_test_split
```

## Importing Data

```
data_Taiwan = pd.read_csv('/content/gdrive/My Drive/Colab/Dissertação/data/data_know.csv').drop("Unnamed: 0", axis=1)
```

```
data_Taiwan.head()
```

## Dimensionality of Data

```
data_Taiwan.shape
```

## Separation of data into X and Y

```
data_y = data_Taiwan['DEFAULT']
data_x = data_Taiwan.drop("DEFAULT", axis=1)
```

```
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.3, random_state = 10)
print("Description    ","Count X       ","Count Y","                 Porcentagem")
print("Train Data     ",x_train.shape,"  ", y_train.shape,"               ",x_train.shape[0]/data_Taiwan.shape[0]*100,"%")
print("Test Data      ",x_test.shape,"   ", y_test.shape,"                ",x_test.shape[0]/data_Taiwan.shape[0]*100,"%")
print("__________________________________________________________________________")
print("Total          ",data_Taiwan.shape,"                            100%")
```

## Merging data to save

```
data_train = pd.merge(x_train, y_train, on = x_train.ID, how = 'outer')
data_train = data_train.drop(columns = ['key_0'])
data_train.columns
```

```
data_test = pd.merge(x_test, y_test, on = x_test.ID, how = 'outer')
data_test = data_test.drop(columns = ['key_0'])
data_test.columns
```

## Saving Data

```
data_train.to_csv('/content/gdrive/My Drive/Colab/Dissertação/data/data_train.csv')
data_test.to_csv('/content/gdrive/My Drive/Colab/Dissertação/data/data_test.csv')
```
