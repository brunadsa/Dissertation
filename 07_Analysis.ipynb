# Calculating metrics mean

```
def metrics_statistic_mean(dataset, dataset_name, metric_name, metric_mean_name):
  print("Calculating "+ metric_name + " on dataset " + dataset_name)
  for instance in range(dataset.shape[0]):
    metrics_list_str = dataset.iloc[instance][metric_name].replace('[','').replace(']','').replace(' ', '').split(',')
    metrics_list_float = [float(i) for i in metrics_list_str]
    last_index = int(len(metrics_list_float)*0.05)
    metric_value = np.mean(metrics_list_float[-last_index:])
    dataset.loc[instance, metric_mean_name] = metric_value
    
  return dataset
```

```
def metrics_statistic_last(dataset, dataset_name, metric_name, metric_mean_name):
  print("Calculating "+ metric_name + " on dataset " + dataset_name)
  for instance in range(dataset.shape[0]):
    metrics_list_str = dataset.iloc[instance][metric_name].replace('[','').replace(']','').replace(' ', '').split(',')
    metrics_list_float = [float(i) for i in metrics_list_str]
    metric_value = metrics_list_float[-1]
    dataset.loc[instance, metric_mean_name] = metric_value
    
  return dataset
```

# Merge Datasets

```
data_metrics.head()
```

```
frames = [data_features_5, data_features_50, data_features_100]
data_training_all = pd.concat(frames, ignore_index=True)
```

```
weights_list = [0.5, 0.7, 0.8, 0.9]
flip_y_list = [0.01, 0.5, 0.8]
n_features = [5, 50, 100]
```

```
parameters_list = []
for weight in weights_list:
    for flip_y in flip_y_list:
        for n_feature in n_features:
            parameters_ = { 'n_samples' : 10000,
                  'n_features' : n_feature,
                  'n_informative' : 2,
                  'n_redundant'  : 2,
                  'random_state' : 2,
                  'n_classes' : 2,
                  'n_clusters_per_class' : 2,
                  'weights' : (weight,),
                  'class_sep' : 1,
                  'flip_y' : flip_y,
                  'hypercube': True,
                  'shuffle': True}
            parameters_list.append(parameters_)
```

```
data_training_all['weights'] = -1
data_training_all['flip_y'] = -1
data_training_all['n_features'] = -1
```

```
columns_ = ['dataset','layers','iteration','features', 'weights', 'flip_y','training_accuracy','validation_accuracy','accuracy_test']
```

```
data_training_all.loc[0, 'C1_metrics'] = data_metrics['C1_metrics'][0]
```

```
data_metrics.head()
```

```
for parameters_index in range(len(parameters_list)):
  index_on_data = parameters_index + 1
  print("Dataset: ", index_on_data)
  row_indexer_list = data_training_all[data_training_all['dataset'] == index_on_data].index
  print("Index from dataset:", row_indexer_list)
  for row_indexer in row_indexer_list:
    data_training_all.loc[row_indexer, 'flip_y'] = parameters_list[parameters_index]['flip_y']
    data_training_all.loc[row_indexer, 'weights'] = parameters_list[parameters_index]['weights'][0]
    data_training_all.loc[row_indexer, 'n_features'] = parameters_list[parameters_index]['n_features']

    data_training_all.loc[row_indexer, 'C1_metrics'] = data_metrics['C1_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'C2_metrics'] = data_metrics['C2_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'C3_metrics'] = data_metrics['C3_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'C4_metrics'] = data_metrics['C4_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'L1_metrics'] = data_metrics['L1_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'L2_metrics'] = data_metrics['L2_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'L3_metrics'] = data_metrics['L3_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'S1_metrics'] = data_metrics['S1_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'S2_metrics'] = data_metrics['S2_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'S3_metrics'] = data_metrics['S3_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'S4_metrics'] = data_metrics['S4_metrics'][parameters_index]
    data_training_all.loc[row_indexer, 'T2_metrics'] = data_metrics['T2_metrics'][parameters_index]
```
